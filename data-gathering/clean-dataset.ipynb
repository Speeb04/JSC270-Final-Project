# Merging The Original Datasets

```python
import pandas as pd

df1 = pd.read_csv("../data/new-dataset.csv")  # contains 'track_id'
df2 = pd.read_csv("../data/original-dataset.csv")  # contains 'spotify_id'
```

Merging on the left, based on the Spotify ID

```python
merged = pd.merge(
    df1,
    df2,
    left_on="track_id",
    right_on="spotify_id",
    how="inner"
)
```

```python
merged
```

Dropping NA for empty genres

```python
merged.dropna(subset=["genre"], inplace=True)
```

```python
merged
```

Remove unused columns

```python
merged.drop(columns=['Unnamed: 0'], inplace=True)
```

```python
merged
```

```python
merged.to_csv("output.csv")
```
